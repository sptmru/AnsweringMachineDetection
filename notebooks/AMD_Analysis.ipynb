```python
import glob
import os
import librosa
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from matplotlib.pyplot import specgram
%matplotlib inline

def plot_specgram(file_path):
  y, sr = librosa.load(file_path)
  S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128,fmax=8000)
  plt.figure(figsize=(10, 4))
  librosa.display.specshow(librosa.power_to_db(S,ref=np.max),y_axis='mel', fmax=8000,x_axis='time')
  plt.colorbar(format='%+2.0f dB')
  plt.title(file_path.split("/")[-2])
  plt.tight_layout()

sound_file_paths = ["/Users/sptm/Desktop/amd-dataset-stuff/all-recordings/beep_1500_0.2_1.0.wav",
                    "/Users/sptm/Desktop/amd-dataset-stuff/all-recordings/overdrive_beep_1500_0.2_1.5.wav",
                    "/Users/sptm/Desktop/amd-dataset-stuff/all-recordings/whitenoise_beep_1500_0.2_1.1.wav",
                    "/Users/sptm/Desktop/amd-dataset-stuff/all-recordings/beep_500_0.2_1.1.wav",
                    "/Users/sptm/Desktop/amd-dataset-stuff/all-recordings/overdrive_beep_500_1_1.4.wav",
                    "/Users/sptm/Desktop/amd-dataset-stuff/all-recordings/whitenoise_beep_500_1_1.4.wav",
                    "/Users/sptm/Desktop/amd-dataset-stuff/all-recordings/silence_1.3.wav",
                    "/Users/sptm/Desktop/amd-dataset-stuff/all-recordings/ad_padded_1.5.wav"
                    ]

for file in sound_file_paths:
  plot_specgram(file)
```
